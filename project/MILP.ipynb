```python
init_state=4
bat_state=[0.5*init_state]#要输出的电池状态，这是0时刻的初始状态
pg_out=[0]#对应着pg+，输出
pg_in=[0]#对应着pg-，输出
pb_in=[0]#对应着pb+，输出
pb_out=[0]#对应着pb-，输出
my_ub =[float(np.max(Total_Load_array)) for i in range(96)]#,4 for i in range(96),1,1
for i in range(96):
    my_ub.append(4.0)
my_ub.append(1.0)
my_ub.append(1.0)
my_ub.append(4.0)#用户47的电池最大容量
my_lb=[0 for i in range(195)]

my_ctype=[]#对每个参数类型进行定义，I为Integer，C为Constant
for i in range(192):
    my_ctype.append('C')
my_ctype.append('I')
my_ctype.append('I')
my_ctype.append('C')

my_colnames=[]#所有变量的名称，按照pg+,pg-,pb+,pb-,d,s,e这个顺序
for i in range(48):
    my_colnames.append("pg+"+str(i))
for i in range(48):
    my_colnames.append("pg-"+str(i))
for i in range(48):
    my_colnames.append("pb+"+str(i))
for i in range(48):
    my_colnames.append("pb-"+str(i))
my_colnames.append("d")
my_colnames.append("s")
my_colnames.append("e_b")


my_rownames =[]#约束条件的行数的名称
for i in range(6):
    my_rownames .append("r"+str(i+1))

my_sense ="EELLLL"


row1=[]#第一个约束条件的系数
for i in range(48):
    row1.append(1.0)
for i in range(48):
    row1.append(-1.0)
for i in range(48):
    row1.append(-0.819)
for i in range(48):
    row1.append(0.989)
for i in range(3):
    row1.append(0.0)
row3=[]#第三个约束条件的系数
for i in range(48):
    row3.append(1.0)
for i in range(144):
    row3.append(0.0)
row3.append(-float(np.max(Total_Load_array)))
row3.append(0.0)
row3.append(0.0)

row4=[]#第四个
for i in range(48):
    row4.append(0.0)
for i in range(48):
    row4.append(1.0)
for i in range(96):
    row4.append(0.0)    
row4.append(float(np.max(Total_Load_array)))
row4.append(0.0)
row4.append(0.0)

row5=[]#第五个约束条件的系数
for i in range(96):
    row5.append(0.0)
for i in range(48):
    row5.append(1.0)
for i in range(49):
    row5.append(0.0)
row5.append(-4.0)
row5.append(0.0)

row6=[]#第六个约束条件的系数
for i in range(144):
    row6.append(0.0)
for i in range(48):
    row6.append(1.0)
row6.append(0.0)
row6.append(4.0)
row6.append(0.0)
len(row6)
```

```python
for i in range(48):#对一天时间范围内进行滚动求解
    my_rhs = [float(Total_Load_array[1][i]-0.9*GG_array[1][i]),
              float(bat_state[-1]+0.455*pb_in[-1]-(0.5/0.91)*pb_out[-1]),
              0.0,float(np.max(Total_Load_array)),0.0,4.0]#约束条件等式的右侧值，是变化的
    tou=[]
    tou.extend(T_tou[i:])
    tou.extend(T_tou[0:i])#输入分时电价
    for j in range(48):
        tou.append(T_fit)
    for i in range(99):
        tou.append(0)#到此处就建立了一天的所有参数的系数，共有48*4+2+1=195个，包含两个01决策变量
    print(len(tou))
    mlp=cplex.Cplex()
    mlp.objective.set_sense(mlp.objective.sense.minimize)
    mlp.variables.add(obj=tou,
		    lb=my_lb,
		    ub=my_ub,
		    types=my_ctype,
		    names=my_colnames)
    rows = [[my_colnames,row1],#第一个约束条件
    	    [["e_b"],[1]],
    	    [my_colnames,row3],[my_colnames,row4],[my_colnames,row5],
            [my_colnames,row6]]
    mlp.linear_constraints.add(lin_expr=rows,
    				senses=my_sense,
    				rhs=my_rhs,
    				names=my_rownames)
    mlp.solve()
    x = mlp.solution.get_values()
    bat_state.append(x[-1])
    pg_out.append(x[0])#对应着pg+，输出
    pg_in.append(x[48])#对应着pg-，输出
    pb_in.append(x[96])#对应着pb+，输出
    pb_out.append(x[144])#对应着pb-，输出
    tou=[]#为了组合的大小始终为195所必须的操作
```
